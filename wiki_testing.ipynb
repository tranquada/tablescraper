## Key variable declarations

```python
url = "https://survivingmars.gamepedia.com/Buildings" # "Buildings" page from the Surviving Mars wiki
```

## Page exploration with Beautiful Soup

```python
page = r.get(url)
wiki = s(page.text, "lxml")
```

```python
wiki
```

```python
titles = wiki.find_all("h2")[1:]
tables = wiki.find_all("table")
```
